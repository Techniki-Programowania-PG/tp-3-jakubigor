# Wizualizacja sygna≈Çu z wykorzystaniem biblioteki matplotplusplus

```python
a = a3.cos(2, 0, 10, 100)
b = a3.sin(0.5, 0, 10, 100)
c = a3.square(1, 0, 10, 100)
d = a3.sawtooth(1, 0, 10, 100)
```

```python
a3.plot(a)
Image(filename='plot.png')
```

# DFT i transformata odwrotna

```python
a3.cplot(a3.dft(a))
Image(filename='plot.png')
```

```python
a3.plot(a3.idft(a3.dft(a)))
Image(filename='plot.png')
```

# Filtracja 1D i 2D

```python
a3.plot(a3.filter1d(a, 0.01))
Image(filename='plot.png')
```

```python
a3.cplot(a3.filter2d(a3.dft(b), 3))
Image(filename='plot.png')
```
